```python
import time
import numpy as np


sizes = [100000, 1000000, 10000000, 100000000]


def test_operations():
    for size in sizes:
        print(f"\nТестирование с {size} элементами:")


        python_list = list(range(size))
        numpy_array = np.arange(size)

        start_time = time.time()
        sum_python = sum(python_list)
        time_python = time.time() - start_time

        start_time = time.time()
        sum_numpy = np.sum(numpy_array)
        time_numpy = time.time() - start_time

        print(f"Сумма: Python = {time_python:.6f} сек, NumPy = {time_numpy:.6f} сек")

        start_time = time.time()
        avg_python = sum(python_list) / len(python_list)
        time_python = time.time() - start_time

        start_time = time.time()
        avg_numpy = np.mean(numpy_array)
        time_numpy = time.time() - start_time

        print(f"Среднее: Python = {time_python:.6f} сек, NumPy = {time_numpy:.6f} сек")

        start_time = time.time()
        multiplied_python = [x * 2 for x in python_list]
        time_python = time.time() - start_time

        start_time = time.time()
        multiplied_numpy = numpy_array * 2
        time_numpy = time.time() - start_time

        print(f"Умножение: Python = {time_python:.6f} сек, NumPy = {time_numpy:.6f} сек")

test_operations()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')

print("верхний левый угол:")
x_min, y_max = map(int, input().split())
print("правый нижний угол:")
x_max, y_min = map(int, input().split())

filtered_df = df[(df['x'] >= x_min) & (df['x'] <= x_max) &
                 (df['y'] >= y_min) & (df['y'] <= y_max)]

plt.figure(figsize=(8, 6))
plt.scatter(filtered_df['x'], filtered_df['y'], color='blue', s=5, label='Выстрелы')
plt.title('Координаты выстрелов в заданном прямоугольнике')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.grid(True)
plt.legend()
plt.show()
```
